```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_fullsent = pd.read_csv('fulldata_sentiment_analysis_results.csv')
df_fullsent.sample(20)
```

```python

# Filter the data where the priority is HIGH
high_priority_data = df_fullsent[df_fullsent['priority'] == 'HIGH']

high_priority_data.info()
```

```python
high_priority_data 
```

```python
high_negative_tickets = df_fullsent[(df_fullsent['priority'] == 'HIGH') & (df_fullsent['sentiment_label'] == 'negative')]

high_negative_tickets
```

```python
# Order the filtered data by sentiment score from high to low
ordered_by_score = high_negative_tickets.sort_values(by='sentiment_score', ascending=False)

ordered_by_score
```

```python
ordered_by_score["text"]
```

```python
# Display the first few rows of the ordered filtered data with the 'text' column
display_columns = ['queue', 'priority', 'language', 'subcategory', 'subject', 'text', 'sentiment_label', 'sentiment_score']
ordered_filtered_data_display = ordered_by_score[display_columns]

ordered_filtered_data_display
```

```python
# Get the row with the highest sentiment score for high priority and negative sentiment label
highest_sentiment_score_row = ordered_filtered_data_display.iloc[0]

highest_sentiment_score_row
```

```python
# Extract the text from that row
highest_sentiment_score_text = highest_sentiment_score_row['text']

highest_sentiment_score_text 
```

```python
import random


# Define average handling time and number of agents
average_handling_time = 20  # in minutes
number_of_agents = 10

# Assign tickets to agents randomly
tickets = list(high_priority_data['text'])
random.shuffle(tickets)

# Find the position of the specific ticket
specific_ticket = "Our subscription system is completely unresponsive. We've tried restarting the server but no luck. Need urgent help!"
ticket_position = tickets.index(specific_ticket)

# Calculate the time taken to reach this ticket
# Divide the position by the number of agents to simulate concurrent handling
time_to_settle = (ticket_position // number_of_agents + 1) * average_handling_time

# Convert time to hours and minutes
hours, minutes = divmod(time_to_settle, 60)

# Display the result
print(f"The ticket '{specific_ticket}' will be settled in approximately {hours} hours and {minutes} minutes.")
```
